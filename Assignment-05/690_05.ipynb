**Calculate the min, max, and average in-state tution of the above four institutions.**

```
list_four_colleges = []
```

```
df_1_fee = df_2.loc[df_2['INSTNM'] == 'University of Maryland-Baltimore County', 'TUITIONFEE_IN'].iloc[0]
list_four_colleges.append(df_1_fee)
```

```
df_2_fee = df_2.loc[df_2['INSTNM'] == 'University of Maryland-College Park', 'TUITIONFEE_IN'].iloc[0]
list_four_colleges.append(df_2_fee)
```

```
df_3_fee = df_2.loc[df_2['INSTNM'] == 'Loyola University Maryland', 'TUITIONFEE_IN'].iloc[0]
list_four_colleges.append(df_3_fee)
```

```
df_4_fee = df_2.loc[df_2['INSTNM'] == 'Johns Hopkins University', 'TUITIONFEE_IN'].iloc[0]
list_four_colleges.append(df_4_fee)
```

```
list_four_colleges
```

**Min Value in Four Colleges**

```
min(list_four_colleges)
```

**Max Value in Four Colleges**

```
max(list_four_colleges)
```

**Mean Value of Four Colleges in-state tution fees**

```
sum = 0
for i in list_four_colleges:
  sum+= i
print(sum)
```

```
mean = sum / len(list_four_colleges)
print("The mean in-state tution fee value for mean", mean)
```
