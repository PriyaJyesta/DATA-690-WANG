**As the columns names are not correct, cleaning the dataset to display headers**

```
df1.columns = df1.iloc[0]
df1 = df1.drop(0)
df1.head()
```

**Function to cleanse the Census Year column : by remove "*" in year column data**

```
def cleanse_year(raw_cost):
    raw_cost = int (raw_cost.strip('*'))
    return raw_cost
```

```
df1["Census Year"] = df1["Census Year"].apply(cleanse_year)
df1
```

```
df1["Total Population"] = df1["Total Population"].astype(int)
```

**Function to cleanse the Census Year column : by remove "$" and "Billion" in Census Cost column data to make uniform measurement unit throguhtout the column's data**

```
def cleanse_cost(raw_cost):
    cost = raw_cost.lstrip("$").replace(",","")
    if  raw_cost.find("4.5") != -1:          
        cost = "45"+"0" * 8
    elif raw_cost.find("Billion") != -1:     
        cost = "13"+"0" * 9
    return int (cost)
```

```
df1["Census Cost"] = df1["Census Cost"].apply(cleanse_cost)
df1
```

```
def cleanse_cent(raw_cost):
    if  raw_cost.find("cent") != -1:
        raw_cost = raw_cost.replace("cents", "").strip()           
        raw_cost = float(raw_cost) * 0.01
    elif raw_cost.find("$") != -1:
        raw_cost = float(raw_cost.replace("$", "").strip())     
    return raw_cost
```

```
df1["Average Cost Per Person"] = df1["Average Cost Per Person"].apply(cleanse_cent)
df1
```
